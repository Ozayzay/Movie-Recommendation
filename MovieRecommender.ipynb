### Fix genres Column

```python
#We can see the genres are in a very weird format we want to bring it into this format down below:
# ['Action','Adventure',Fantasy','SciFi']
movies.iloc[0].genres
```

```python
# We do this using a helper function to extract name but we have a problem that this genres columns values are a string and
# not a dictionary that we would want so we have to convert this string of lists into lists we do that using ast literal


import ast


def convert(obj):
    List = []
    # now we take one dictionary list at a time and extract the name from it 
    for i in ast.literal_eval(obj):
        List.append(i["name"])
    return List


movies["genres"] = movies["genres"].apply(convert)
movies.head()
    
```

### Fix Keywords 

```python
movies["keywords"] = movies["keywords"].apply(convert)
movies.head()
```

### Fix Cast 
Only keep first 3 dictionaries 

```python
#movies["cast"][0]


def convert3(obj):
    List = []
    counter= 0
    # now we take one dictionary list at a time and extract the name from it 
    for i in ast.literal_eval(obj):
        if counter !=3:
            List.append(i["name"])
            counter+=1
        else:
            break
    return List


movies["cast"] = movies["cast"].apply(convert3)
movies.head()
```

### Fix Crew 

```python
#movies["crew"][0]
#{"credit_id": "52fe48009251416c750ac9c3", "department": "Directing", "gender": 2, "id": 2710, "job": "Director", "name": "James Cameron"}, 
```

```python
def convertdirector(obj):
    List = []
    # now we take one dictionary list at a time and extract the name from it 
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            List.append(i["name"])
            break
    return List
```

```python
movies["crew"]= movies["crew"].apply(convertdirector)
movies.head()
```

### Converting Movies Overview column into a list so we can concatinate it with the other lists 

```python
# Splits the rows in overview column with a space as the default delimiter 
movies["overview"]= movies["overview"].apply((lambda x:x.split()))
movies.head()
```

### For the genres , keywords , cast and crew columns if the keywords have a space in them we will remove it

Reason for this is if you leave in the spaces - for example the name " Same Mendes"  our process will make this into two seperate tags of sam and mendes and won't think of this as one entity 

```python
# Replace spaces in lists of strings for the specified columns
columns_to_clean = ["genres", "keywords", "cast", "crew"]

def remove_spaces_from_values(column_values):
    # Process each value in the list
    return [value.replace(" ", "") for value in column_values]

# Apply the function to each specified column
for column in columns_to_clean:
    movies[column] = movies[column].apply(remove_spaces_from_values)

# Display the DataFrame to verify changes
movies.head()
```

### Make a new column called tags which concats all these 5 columns 

```python
movies['tags'] = movies['overview'] + movies['genres']+ movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
movies = movies[["movie_id","title","tags"]]
movies
```

```python
# Convert the tags list to a string 
# ' '.join(x) combines all elements of the list x into a single string, separated by spaces.
# word.strip() Removes leading and trailing whitespace from each word in the list.
# Generator expression : word.strip() for word in x processes each element in the list x to clean up spaces.

# Remove extra spaces from each word and join the list into a single string
movies['tags'] = movies['tags'].apply(
    lambda x: ' '.join(word.strip() for word in x) if isinstance(x, list) else x
)

print(movies['tags'][0])
```
